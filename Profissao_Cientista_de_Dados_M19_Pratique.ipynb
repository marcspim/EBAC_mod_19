**Efetividade de Duas Estratégias de Ensino**

Imagine que uma escola esteja avaliando a eficácia de duas estratégias de ensino de matemática para alunos do ensino médio. Eles querem determinar se há uma diferença significativa no desempenho médio dos alunos entre as duas estratégias.

# **Hipóteses:**

* Hipótese nula (H0): A média das notas dos alunos na estratégia A é igual à média das notas dos alunos na estratégia B.
* Hipótese alternativa (H1): A média das notas na Estratégia B é maior do que a média das notas na Estratégia A.

# **Dados:**

* Amostra da Estratégia A: Notas de 50 alunos que receberam a Estratégia A.
* Amostra da Estratégia B: Notas de 50 alunos que receberam a Estratégia B.

Usaremos um teste Z para comparar as médias das notas entre as duas amostras.

Se o p-valor do teste Z for menor que um nível de significância pré-determinado (por exemplo, α = 0.05), rejeitamos a hipótese nula e concluímos que há uma diferença significativa nas médias das notas entre as duas estratégias de ensino.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

Os dados são criados a seguir:

```python
# Definindo médias e desvios padrão para as notas nas duas estratégias
media_estrategia_A = 70
desvio_padrao_estrategia_A = 10

media_estrategia_B = 75
desvio_padrao_estrategia_B = 12

# Gerando as amostras de notas para cada estratégia de ensino da nossa base
np.random.seed(0)  # Para garantir a reprodutibilidade dos resultados
amostra_estrategia_A = np.random.normal(loc=media_estrategia_A, scale=desvio_padrao_estrategia_A, size=50)
amostra_estrategia_B = np.random.normal(loc=media_estrategia_B, scale=desvio_padrao_estrategia_B, size=50)

print("Notas da Estratégia A:", amostra_estrategia_A[:5])
print("Notas da Estratégia B:", amostra_estrategia_B[:5])
```

# 1) De acordo com as informações analisadas o nosso teste é unilateral á direita, esquerda ou bicaudal? Justifique.




```python
# Sua justificativa aqui
```

O teste é unilateral à direita, pois a hipótese alternativa (H1) afirma que a média das notas dos alunos que utilizaram a Estratégia B é maior do que a média das notas dos alunos da Estratégia A (µB > µA), ou seja, queremos verificar se a média da Estratégia B é superior à da Estratégia A, o que indica que a região crítica do teste (onde rejeitamos H0) estará localizada apenas na cauda direita da distribuição da estatística do teste Z.

# 2) Calcule as médias para as duas amostragens e as variâncias. Quais insights você pode retirar comparando os dados?

```python
#Seu código aqui
# Calculando as médias amostrais
media_amostral_A = np.mean(amostra_estrategia_A)
media_amostral_B = np.mean(amostra_estrategia_B)

print(f"Média amostral A: {media_amostral_A:.2f}")
print(f"Média amostral B: {media_amostral_B:.2f}")

# Calculando as variâncias amostrais
var_A = amostra_estrategia_A.var(ddof=1)
var_B = amostra_estrategia_B.var(ddof=1)

print(f"Variância de A: {var_A}")
print(f"Variância de B: {var_B}")
```

A média da Estratégia B é maior que a da Estratégia A, o que nos leva a tender para o aceite da hipótese alternativa.

A variância da Estratégia A é maior do que a da Estratégia B, sugerindo que as notas dos alunos na Estratégia A estão mais dispersas em relação à média, havendo maior variabilidade no desempenho dos alunos que usaram a Estratégia A.

Portanto, a Estratégia B parece oferecer um desempenho mais consistente aos alunos, sendo uma informação relevante para análise pedagógica, visto que uma menor dispersão pode significar que a estratégia gerou resultados mais homogêneos entre os alunos.

# 3) Imprima os resultados da estatística do teste Z, p value e indique se rejeitaremos ou não a hipótese nula. Justifique sua resposta.

```python
#Seu código aqui
# Tamanho das amostras
n = len(amostra_estrategia_A)

# Calculando o valor do teste Z
z_teste = (media_amostral_B - media_amostral_A) / ((var_A/n + var_B/n)**0.5)

# Nível de significância (alpha)
alpha = 0.05

# Calculando o p-valor
p_value = 1 - stats.norm.cdf(z_teste)

# Exibindo os resultados
print(f"Valor do teste Z: {z_teste:.4f}")
print(f"p-value: {p_value:.4f}")

if p_value < alpha:
    print("Rejeitamos H0: A média das notas na Estratégia B é significativamente maior que na Estratégia A.")
else:
    print("Falhamos em rejeitar H0: Não há evidências suficientes para afirmar que a Estratégia B tem média superior.")
```

Como o p-value é superior ao alpha determinado para o nosso teste Z, a distribuição dos nossos dados e os resultados da estatística aplicada nos mesmos não conseguem providenciar evidência que sejam suficientes para que, a nível de significância estatística, possamos afirmar com embasamento matemático que a Estratégia B forneceu aos alunos uma média superior àqueles que utilizaram da Estratégia A.

# 4) Para finalizar monte o gráfico da distribuição da estatística do nosso teste Z e explique o que pode ser observado através dele.

```python
#Seu código aqui
# Configurações do gráfico
fig, ax = plt.subplots(figsize=(8, 5))

# Valores de X e Y para plotar a distribuição
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

# Plotando a distribuição normal padrão
plt.plot(x, y, label="Distribuição Normal Padrão", color='blue')

# Marcação do valor observado da estatística do teste
plt.axvline(z_teste, color='red', linestyle='--', label=f'Z teste = {z_teste:.2f}')

# Área crítica
# Obtendo o valor crítico para um teste unilateral à direita
z_critico = stats.norm.ppf(1 - alpha)
plt.axvline(z_critico, color='green', linestyle='--', label=f'Z crítico = {z_critico:.2f}')
plt.fill_between(x, y, where=(x >= z_critico), color='red', alpha=0.3, label='Região crítica')

# Legenda e rótulos
plt.legend()
plt.xlabel('Valor da Estatística de Teste Z')
plt.ylabel('Densidade')
plt.title('Distribuição da Estatística de Teste Z')
plt.grid(True)

# Exibindo o gráfico
plt.show()
```

A curva azul representa a distribuição normal padrão, com média 0 e desvio padrão 1. A linha verde marca o valor crítico, que define a região de rejeição de H0. A linha vermelha representa o valor da estatística do teste Z, que indica a posição da média amostral da Estratégia B em relação à distribuição esperada sob H0. A região vermelha à direita de Z crítico representa a região crítica onde rejeitaríamos H0 caso o teste Z estivesse nela.

Como o valor do teste Z não está na região crítica, não rejeitamos H0 ao nível de significância de 5%, o que confirma que não há evidências estatísticas suficientes para afirmar que a Estratégia B tem um desempenho superior à Estratégia A.
